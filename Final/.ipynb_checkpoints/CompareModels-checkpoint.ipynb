# Comparision of the Three Different Neural Networks.

This file is a comparision of the three different neural networks tested, an ordinary neural network, a base model neural network, and an adversarial neural network which uses the base model neural network.

All three models are run for various training data set sizes and the results compared.

```python
from make_data import make_data
from particle_net import particle_net
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras import layers
from mpl_toolkits import mplot3d
import neural_structured_learning as nsl
from collections import defaultdict
from adversary import adversary
from base_model import base_model
```

The user can decide the constant momentum used for this simulation for all particles, as well as the size of the training and testing data sets array.  The actual size of the test and training data sets is doubled as these numbers are for each distribution. 

```python
Epochs = 5
momentum = 100
N = [100,250,500,750,1000,2500,5000,7500,10000,25000,50000,75000,100000,250000,500000,750000,1000000]
```

Here, I run all three models for all N values and store the data for plotting later in this file.

```python
X = len(N)
FalseExpsOridinary = np.zeros((X, 2))
FalseGaussOridinary = np.zeros((X, 2))
FalseExpsBase = np.zeros((X, 2))
FalseGaussBase = np.zeros((X, 2))
FalseExpsAdversary = np.zeros((X, 2))
FalseGaussAdversary = np.zeros((X, 2))
j = 0

for i in N:
    
    Ntrain = i
    Ntest = i
    
    train,train_vals = make_data(momentum,Ntrain)
    test,test_vals = make_data(momentum,Ntest)

    predictions = particle_net(train,train_vals,test,Epochs)
    
    FalseExpsOridinary[j,:] = sum(predictions[np.where(test_vals == 0)])/Ntest*100
    FalseGaussOridinary[j,:] = sum(predictions[np.where(test_vals == 1)])/Ntest*100

    predicts = base_model(train, train_vals, test, Epochs)

    FalseExpsBase[j,:] = sum(predicts[np.where(test_vals == 0)])/Ntest*100
    FalseGaussBase[j,:] = sum(predicts[np.where(test_vals == 1)])/Ntest*100
    
    predict = adversary(train, train_vals, test, test_vals,Epochs)

    FalseExpsAdversary[j,:] = sum(predict[np.where(test_vals == 0)])/Ntest*100
    FalseGaussAdversary[j,:] = sum(predict[np.where(test_vals == 1)])/Ntest*100
    
    j = j + 1
```

This is a plot of the models' performances for incorrect classifications for both classifications.

```python
fig = plt.figure()
plt.plot(N,FalseExpsOridinary[:,1], label = 'Ordinary - Exp')
plt.plot(N,FalseExpsBase[:,1], label = 'Base - Exp')
plt.plot(N,FalseExpsAdversary[:,1], label = 'Adversary - Exp')
plt.plot(N,FalseGaussOridinary[:,0], label = 'Ordinary - Guassian')
plt.plot(N,FalseGaussBase[:,0], label = 'Base - Guassian')
plt.plot(N,FalseGaussAdversary[:,0], label = 'Adversary - Guassian')
plt.xlabel('Size of Training and Testing data sets')
plt.ylabel('Accuracy Percentage')
plt.title('Performance of Models - Incorrect Classifications')
plt.legend()
fig.savefig('CompareIncorrect.png')
```

This is a plot of the models' performances for correct classifications for both classifications.

```python
fig = plt.figure()
plt.plot(N,FalseExpsOridinary[:,0], label = 'Ordinary - Exp')
plt.plot(N,FalseExpsBase[:,0], label = 'Base - Exp')
plt.plot(N,FalseExpsAdversary[:,0], label = 'Adversary - Exp')
plt.plot(N,FalseGaussOridinary[:,1], label = 'Ordinary - Guassian')
plt.plot(N,FalseGaussBase[:,1], label = 'Base - Guassian')
plt.plot(N,FalseGaussAdversary[:,1], label = 'Adversary - Guassian')
plt.xlabel('Size of Training and Testing data sets')
plt.ylabel('Accuracy Percentage')
plt.title('Performance of Models - Correct Classifications')
plt.legend()
fig.savefig('CompareCorrect.png')
```

Below is a plot of the models' performances for both correct and incorrect classifications for the expodential mass classification.

```python
fig = plt.figure()
plt.plot(N,FalseExpsOridinary[:,0], label = 'Ordinary - Correct')
plt.plot(N,FalseExpsBase[:,0], label = 'Base - Correct')
plt.plot(N,FalseExpsAdversary[:,0], label = 'Adversary - Correct')
plt.plot(N,FalseExpsOridinary[:,1], label = 'Ordinary - Wrong')
plt.plot(N,FalseExpsBase[:,1], label = 'Base - Wrong')
plt.plot(N,FalseExpsAdversary[:,1], label = 'Adversary - Wrong')
plt.xlabel('Size of Training and Testing data sets')
plt.ylabel('Accuracy Percentage')
plt.title('Performance of Models for Expodential')
plt.legend()
fig.savefig('CompareExpodential.png')
```

Below is a plot of the models' performances for both correct and incorrect classifications for the Gaussian mass classification.

```python
fig = plt.figure()
plt.plot(N,FalseGaussOridinary[:,1], label = 'Ordinary - Correct')
plt.plot(N,FalseGaussBase[:,1], label = 'Base - Correct')
plt.plot(N,FalseGaussAdversary[:,1], label = 'Adversary - Correct')
plt.plot(N,FalseGaussOridinary[:,0], label = 'Ordinary - Wrong')
plt.plot(N,FalseGaussBase[:,0], label = 'Base - Wrong')
plt.plot(N,FalseGaussAdversary[:,0], label = 'Adversary - Wrong')
plt.xlabel('Size of Training and Testing data sets')
plt.ylabel('Accuracy Percentage')
plt.title('Performance of Models for Gaussian')
plt.legend()
fig.savefig('CompareGaussian.png')
```

Below is a plot of the absolute value difference in the models' performances between the expodential and the gaussian classifications for each model.

```python
fig = plt.figure()
plt.plot(N,abs(FalseExpsOridinary[:,0]-FalseGaussOridinary[:,1]), 'r*', label = 'Ordinary')
plt.plot(N,abs(FalseExpsBase[:,0]-FalseGaussBase[:,1]), 'b*',label = 'Base')
plt.plot(N,abs(FalseExpsAdversary[:,0]-FalseGaussAdversary[:,1]), 'g*',label = 'Adversary')
plt.plot(N,abs(FalseExpsOridinary[:,0]-FalseGaussOridinary[:,1]), 'r')
plt.plot(N,abs(FalseExpsBase[:,0]-FalseGaussBase[:,1]), 'b')
plt.plot(N,abs(FalseExpsAdversary[:,0]-FalseGaussAdversary[:,1]), 'g')
plt.xlabel('Size of Training and Testing data sets')
plt.ylabel('Accuracy Percentage')
plt.title('Performance of Models - Correct Classifications')
plt.legend()
fig.savefig('CompareDifference.png')
```

As can be seen, all models performed reasonably well for larger data sets of 750k and better.
