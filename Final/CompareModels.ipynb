```python
from MakeData import MakeData
from ParticleNet import ParticleNet
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras import layers
from mpl_toolkits import mplot3d
import neural_structured_learning as nsl
from collections import defaultdict
from Adversary import Adversary
```

```python
Epochs = 10
p = 100
N = [100,250,500,750,1000,2500,5000,7500,10000,25000,50000,75000,100000,250000,500000,750000,1000000]
```

```python
X = len(N)
FalseExpsOridinary = np.zeros((X, 2))
FalseGaussOridinary = np.zeros((X, 2))
FalseExpsAdversary = np.zeros((X, 2))
FalseGaussAdversary = np.zeros((X, 2))
j = 0

for i in N:
    
    Ntrain = i
    Ntest = i
    
    train,train_vals = MakeData(p,Ntrain)
    test,test_vals = MakeData(p,Ntest)

    predictions = ParticleNet(train,train_vals,test,Epochs)
    
    FalseExpsOridinary[j,:] = sum(predictions[np.where(test_vals == 0)])/Ntest*100
    FalseGaussOridinary[j,:] = sum(predictions[np.where(test_vals == 1)])/Ntest*100

    predict = Adversary(train, train_vals, test, test_vals,Epochs)

    FalseExpsAdversary[j,:] = sum(predict[np.where(test_vals == 0)])/Ntest*100
    FalseGaussAdversary[j,:] = sum(predict[np.where(test_vals == 1)])/Ntest*100
    
    j = j + 1
```

```python
fig = plt.figure()
plt.plot(N,FalseExpsOridinary[:,1], label = 'Ordinary - Exp')
plt.plot(N,FalseExpsAdversary[:,1], label = 'Adversary - Exp')
plt.plot(N,FalseGaussOridinary[:,0], label = 'Ordinary - Guassian')
plt.plot(N,FalseGaussAdversary[:,0], label = 'Adversary - Guassian')
plt.xlabel('Size of Training and Testing data sets')
plt.ylabel('Accuracy Percentage')
plt.title('Performance of Models - Incorrect Classifications')
plt.legend()
fig.savefig('CompareIncorrect.png')
```

```python
fig = plt.figure()
plt.plot(N,FalseExpsOridinary[:,0], label = 'Ordinary - Exp')
plt.plot(N,FalseExpsAdversary[:,0], label = 'Adversary - Exp')
plt.plot(N,FalseGaussOridinary[:,1], label = 'Ordinary - Guassian')
plt.plot(N,FalseGaussAdversary[:,1], label = 'Adversary - Guassian')
plt.xlabel('Size of Training and Testing data sets')
plt.ylabel('Accuracy Percentage')
plt.title('Performance of Models - Correct Classifications')
plt.legend()
fig.savefig('CompareCorrect.png')
```

```python
fig = plt.figure()
plt.plot(N,FalseExpsOridinary[:,0], label = 'Ordinary - Correct')
plt.plot(N,FalseExpsAdversary[:,0], label = 'Adversary - Correct')
plt.plot(N,FalseExpsOridinary[:,1], label = 'Ordinary - Wrong')
plt.plot(N,FalseExpsAdversary[:,1], label = 'Adversary - Wrong')
plt.xlabel('Size of Training and Testing data sets')
plt.ylabel('Accuracy Percentage')
plt.title('Performance of Models for Expodential')
plt.legend()
fig.savefig('CompareExpodential.png')
```

```python
fig = plt.figure()
plt.plot(N,FalseGaussOridinary[:,1], label = 'Ordinary - Correct')
plt.plot(N,FalseGaussAdversary[:,1], label = 'Adversary - Correct')
plt.plot(N,FalseGaussOridinary[:,0], label = 'Ordinary - Wrong')
plt.plot(N,FalseGaussAdversary[:,0], label = 'Adversary - Wrong')
plt.xlabel('Size of Training and Testing data sets')
plt.ylabel('Accuracy Percentage')
plt.title('Performance of Models for Gaussian')
plt.legend()
fig.savefig('CompareGaussian.png')
```
